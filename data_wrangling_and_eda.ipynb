```python
# this function is used to draw the boxplot and the density for numerical variables
def numerical_plotting(df, col, title, symb):
    fig, ax = plt.subplots(2, 1, figsize=(10,7),gridspec_kw={"height_ratios": (.2, .8)})
    ax[0].set_title(title,fontsize=18)
    sns.boxplot(x=col, data=df, ax=ax[0])
    ax[0].set(yticks=[])
    sns.distplot(df[col],kde=True)
    plt.xticks(rotation=45)
    ax[1].set_xlabel(col, fontsize=16)
    plt.axvline(df[col].mean(), color='darkgreen', linewidth=3, label='mean=' + str(np.round(df[col].mean(),1)) + symb)
    plt.axvline(df[col].median(), color='red', linewidth=3, label='median='+ str(np.round(df[col].median(),1)) + symb)
    plt.axvline(df[col].mode()[0], color='purple', linewidth=3, label='mode='+ str(df[col].mode()[0]) + symb)
    plt.legend(bbox_to_anchor=(1, 1.03), ncol=1, fontsize=17, fancybox=True, shadow=True, frameon=True)
    plt.tight_layout()
    plt.show()
```

```python
# this function is responsible for plotting the total count for each categorie
def count_categorical_plotting(df, col_name, title, rotate):
    fig, ax = plt.subplots(figsize=(10,7))
    ax = sns.countplot(x=col_name, data=df)
    ax.bar_label(container = ax.containers[0])
    ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)
    
    if rotate:
        fig.set_size_inches(15, 11, forward=True)
        plt.xticks(rotation=90,fontsize=12)
        
    ax.set_xlabel(col_name, fontsize=16)
    ax.set_ylabel("count", fontsize=16)
    ax.set_title(title, fontsize = 20)
    plt.show()

# this function is responsible for plotting the average price for each categorie
def mean_categorical_plotting(df, col_name, title):
    fig, ax = plt.subplots(figsize=(10,7))
    ax = sns.barplot(x=df.index,y = "price", data=df)
    ax.bar_label(ax.containers[0])
    ax.set_xticklabels(ax.get_xticklabels())
    ax.set_title(title,fontsize=18)
    plt.show()
    
# this function is responsible for plotting the boxplot to see the distribution
def boxplot_categorical_plotting(df, col_name, title):
    fig, ax = plt.subplots(figsize=(10,7))
    ax = sns.boxplot(x=col_name, y = "price", data=df)
    ax.set_xticklabels(ax.get_xticklabels())
    ax.set_title(title,fontsize=18)
    plt.show()
```

## Now let's see the significance of the correlation

```python
from scipy import stats
```

```python
def eta_square(df,feature_name):
    
    ss_within = 0
    grand_mean = 0
    for name in df[feature_name].unique():
        df_categorie = df[['price']][df[feature_name]==name]
        ss_within += df_categorie.var()*(df_categorie.shape[0]-1)
        grand_mean += df_categorie.mean()
        
    grand_mean /= len(df[feature_name].unique())
    
    ss_between = 0
    for name in df[feature_name].unique():
        df_categorie = df[['price']][df[feature_name]==name]
        ss_between += df_categorie.shape[0]*((df_categorie.mean() - grand_mean)**2)
          
    ss_total = ss_between + ss_within
    print(f"the grand mean is {grand_mean[0]}")
    print(f"ss_within = {ss_within[0]}   ,   ss_between = {ss_between[0]}")
    print(f"the eta squred {(ss_between/ss_total*100)[0]} %")
    
```

```python
def f_critical(df,feature_name):
    k = len(df[feature_name].unique())
    dfn = k-1
    dfd = df.shape[0]-k
    f_critical = stats.f.ppf(1-.005, dfn, dfd)
    print(f"df_within = {dfn}   ,   df_between = {dfd}")
    print(f"f_critical {f_critical}")
```
